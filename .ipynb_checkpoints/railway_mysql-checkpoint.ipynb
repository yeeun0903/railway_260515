```python
import os
from dotenv import load_dotenv  # 추가!

load_dotenv()  # .env 파일 로드 (추가!)

import pandas as pd
import plotly.express as px
from sqlalchemy import create_engine

HOST = os.getenv("DB_HOST")
PORT = os.getenv("DB_PORT")
USER = os.getenv("DB_USER")
PASSWORD = os.getenv("DB_PASSWORD")
DATABASE = os.getenv("DB_NAME")

engine = create_engine(
    f"mysql+pymysql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}"
)

query = """
SELECT *
FROM brands01
"""

raw_df = pd.read_sql(query, engine)
raw_df
```

```python
raw_df.to_csv("backup_brands01.csv", index=False, encoding="utf-8-sig")
```

```python
summary_df = raw_df.groupby("brand_name")["brand_id"].count().reset_index()
summary_df.columns = ["brand_name", "brand_count"]

summary_df
```

```python
summary_df.to_sql(
    name="brand_summary01",   # 새 테이블 이름
    con=engine,               # DB 연결
    if_exists="replace",      # 이미 있으면 덮어쓰기
    index=False               # 인덱스는 저장 안 함
)
```

```python
import subprocess

def run_git_command(command):
    result = subprocess.run(
        command,
        shell=True,
        capture_output=True,
        text=True
    )

    print(result.stdout)

    if result.stderr:
        print(result.stderr)

run_git_command("git status")
run_git_command("git add backup_brands01.csv")
run_git_command('git commit -m "Feat: Add MySQL data backup"')
run_git_command("git push origin main")
```
